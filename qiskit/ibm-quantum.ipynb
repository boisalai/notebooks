### Produits tensoriels

Ce code utilise Qiskit pour créer et afficher différents états quantiques de base (états de qubit). Voici l'explication de chaque état :

#### États de la base computationnelle (Z)

**$|0\rangle$** et **$|1\rangle$** : Ce sont les états de base standards
- $|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$
- $|1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$

#### États de la base X (Hadamard)

**$|+\rangle$** = $\frac{\sqrt{2}}{2}|0\rangle + \frac{\sqrt{2}}{2}|1\rangle$
- Superposition égale avec phase positive
- C'est l'état obtenu en appliquant une porte Hadamard à $|0\rangle$

**$|-\rangle$** = $\frac{\sqrt{2}}{2}|0\rangle - \frac{\sqrt{2}}{2}|1\rangle$
- Superposition égale avec phase négative
- C'est l'état obtenu en appliquant une porte Hadamard à $|1\rangle$

#### États de la base Y (circulaire)

**$|r\rangle$** (right) = $\frac{\sqrt{2}}{2}|0\rangle + \frac{\sqrt{2}i}{2}|1\rangle$
- Superposition avec phase complexe $+i$
- Représente une rotation "droite" sur la sphère de Bloch

**$|l\rangle$** (left) = $\frac{\sqrt{2}}{2}|0\rangle - \frac{\sqrt{2}i}{2}|1\rangle$
- Superposition avec phase complexe $-i$
- Représente une rotation "gauche" sur la sphère de Bloch

Ces six états forment les états propres des trois opérateurs de Pauli ($X$, $Y$, $Z$) et sont essentiels en informatique quantique pour comprendre les différentes bases de mesure possibles.



## Différences observées

### Porte T dans la première image
$$T = \begin{pmatrix} 1 & 0 \\ 0 & \frac{1+i}{\sqrt{2}} \end{pmatrix}$$

### Porte T dans votre notebook
$$T = \begin{pmatrix} 1 & 0 \\ 0 & \frac{\sqrt{2}}{2} + \frac{\sqrt{2}i}{2} \end{pmatrix}$$

## Ce sont en fait identiques!

Ces deux expressions sont mathématiquement équivalentes :

$$\frac{1+i}{\sqrt{2}} = \frac{1}{\sqrt{2}} + \frac{i}{\sqrt{2}} = \frac{\sqrt{2}}{2} + \frac{\sqrt{2}i}{2}$$

Car : $\frac{1}{\sqrt{2}} = \frac{1}{\sqrt{2}} \cdot \frac{\sqrt{2}}{\sqrt{2}} = \frac{\sqrt{2}}{2}$

## Matrice THSH

De même pour THSH, les différences sont simplement des notations équivalentes :

**Première image :**
$$THSH = \begin{pmatrix} \frac{1+i}{2} & \frac{1-i}{2} \\ \frac{1}{\sqrt{2}} & \frac{i}{\sqrt{2}} \end{pmatrix}$$

**Votre notebook :**
$$THSH = \begin{pmatrix} \frac{1}{2} + \frac{i}{2} & \frac{1}{2} - \frac{i}{2} \\ \frac{\sqrt{2}}{2} & \frac{\sqrt{2}i}{2} \end{pmatrix}$$

## Conclusion

Les résultats sont **identiques**, seule la forme de notation diffère. Qiskit affiche les nombres complexes en séparant explicitement les parties réelles et imaginaires, et rationalise les dénominateurs (élimine $\sqrt{2}$ du dénominateur).

## État de Bell 



$$
\begin{align}
\ket{\pi_0} &= \ket{0}\ket{0} \\
\ket{\pi_1} &= \ket{0}\ket{+} = \frac{1}{\sqrt{2}}\ket{00} + \frac{1}{\sqrt{2}}\ket{01} \\
\ket{\pi_2} &= \frac{1}{\sqrt{2}}\ket{00} + \frac{1}{\sqrt{2}}\ket{11} = \ket{\Phi^+}
\end{align}
$$

On applique une porte de Hadamard (H) sur le **deuxième qubit** pour passer de $\ket{\pi_0}$ à $\ket{\pi_1}$.

Voici la transformation :

$$\ket{\pi_0} = \ket{0}\ket{0}$$

En appliquant $H$ sur le deuxième qubit :

$$H\ket{0} = \ket{+} = \frac{1}{\sqrt{2}}(\ket{0} + \ket{1})$$

On obtient donc :

$$\ket{\pi_1} = \ket{0}\ket{+} = \ket{0} \otimes \frac{1}{\sqrt{2}}(\ket{0} + \ket{1}) = \frac{1}{\sqrt{2}}\ket{00} + \frac{1}{\sqrt{2}}\ket{01}$$

Ensuite, on applique une porte CNOT où **le deuxième qubit est le contrôle** et **le premier qubit est la cible**.

La porte CNOT avec le qubit 2 comme contrôle et le qubit 1 comme cible effectue l'opération suivante :
- Si le qubit de contrôle (qubit 2) est dans l'état $\ket{0}$, le qubit cible (qubit 1) reste inchangé
- Si le qubit de contrôle (qubit 2) est dans l'état $\ket{1}$, le qubit cible (qubit 1) est inversé (flip)

Premier terme : $\ket{00}$
- Qubit de contrôle (qubit 2) : $\ket{0}$
- Pas de changement sur le qubit cible
- Résultat : $\ket{00}$

Deuxième terme : $\ket{01}$
- Qubit de contrôle (qubit 2) : $\ket{1}$
- On inverse le qubit cible (qubit 1) : $\ket{0} \rightarrow \ket{1}$
- Résultat : $\ket{11}$

L'état après l'application de la porte CNOT est donc :

$$\ket{\pi_2} = \frac{1}{\sqrt{2}}\ket{00} + \frac{1}{\sqrt{2}}\ket{11} = \ket{\Phi^+}$$

On obtient ainsi l'état de Bell $\ket{\Phi^+}$, qui est un état intriqué maximalement.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np

# Créer un circuit avec 2 qubits
qc = QuantumCircuit(2)

# Appliquer la porte Hadamard sur le qubit 1 (Y)
qc.h(1)

# Appliquer la porte CNOT avec contrôle sur qubit 1 (Y) et cible sur qubit 0 (X)
qc.cx(1, 0)

# Afficher le circuit
print(qc.draw(reverse_bits=True))

# Afficher le circuit
print("Circuit de Bell:")
display(qc.draw('mpl', reverse_bits=True))

# Tester les 4 états de base et leurs transformations
base_states = ['00', '01', '10', '11']

print("\nTransformations des états de base:")
for state in base_states:
    # Créer un circuit pour chaque état initial
    qc_test = QuantumCircuit(2)
    
    # Initialiser l'état de base
    if state[0] == '1':
        qc_test.x(0)  # Flip le premier qubit (bas)
    if state[1] == '1':
        qc_test.x(1)  # Flip le deuxième qubit (haut)
    
    # Appliquer les portes de Bell
    qc_test.h(1)
    qc_test.cx(1, 0)
    
    # Calculer le vecteur d'état final
    final_state = Statevector.from_instruction(qc_test)
    display(final_state.draw('latex'))
```

## L'enchevêtrement en action 

### Teleportation protocol

$$
\ket{\pi_0} = \ket{\Phi^+}(\alpha\ket{0} + \beta\ket{1})
$$

L'état de Bell est :

$$\ket{\Phi^+} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11})$$

En substituant dans l'expression initiale :

$$\ket{\pi_0} = \frac{1}{\sqrt{2}}(\ket{00} + \ket{11}) \otimes (\alpha\ket{0} + \beta\ket{1})$$

Par la propriété de distributivité du produit tensoriel :

$$\ket{\pi_0} = \frac{1}{\sqrt{2}}\left[\ket{00} \otimes (\alpha\ket{0} + \beta\ket{1}) + \ket{11} \otimes (\alpha\ket{0} + \beta\ket{1})\right]$$

Développons chaque terme :

**Premier terme :**
$$\ket{00} \otimes (\alpha\ket{0} + \beta\ket{1}) = \alpha\ket{00} \otimes \ket{0} + \beta\ket{00} \otimes \ket{1} = \alpha\ket{000} + \beta\ket{001}$$

**Deuxième terme :**
$$\ket{11} \otimes (\alpha\ket{0} + \beta\ket{1}) = \alpha\ket{11} \otimes \ket{0} + \beta\ket{11} \otimes \ket{1} = \alpha\ket{110} + \beta\ket{111}$$

**Résultat final :**

$$\ket{\pi_0} = \frac{1}{\sqrt{2}}\left[\alpha\ket{000} + \beta\ket{001} + \alpha\ket{110} + \beta\ket{111}\right]$$

En réarrangeant les termes :

$$\ket{\pi_0} = \frac{\alpha\ket{000} + \alpha\ket{110} + \beta\ket{001} + \beta\ket{111}}{\sqrt{2}}$$



```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3, 2)

# ====================================
# PRÉPARATION (pas partie du protocole)
# ====================================
# Cette H prépare l'état |ψ⟩ = |+⟩ qu'on veut téléporter
qc.h(0)  # |0⟩ → |+⟩
qc.barrier(label='État initial |ψ⟩=|+⟩')


state = Statevector.from_instruction(qc)
display(state.draw('latex'))


# ====================================
# CRÉATION PAIRE INTRIQUÉE |Φ⁺⟩
# ====================================
qc.h(1)
qc.cx(1, 2)
qc.barrier(label='Paire Bell')

#state = Statevector.from_instruction(qc)
#display(state.draw('latex'))

# ====================================
# PROTOCOLE DE TÉLÉPORTATION
# ====================================
# Étape 1 : CNOT(Q, A)
qc.cx(0, 1)

# Étape 2 : H(Q) ← Cette H fait partie du protocole
qc.h(0)
qc.barrier(label='Alice')

# Étape 3 : Mesures
qc.measure(1, 0)  # A → b
qc.measure(0, 1)  # Q → a
qc.barrier(label='Mesures')

# Étapes 5.1 et 5.2 : Corrections
with qc.if_test((1, 1)):  # Si a = 1
    qc.x(2)

with qc.if_test((0, 1)):  # Si b = 1
    qc.z(2)

print(qc.draw())
```

Voici un circuit quantique qui met en œuvre le protocole de téléportation.

Voir https://quantum.cloud.ibm.com/learning/fr/courses/basics-of-quantum-information/entanglement-in-action/qiskit-implementation

```python
qubit = QuantumRegister(1, "Q")
ebit0 = QuantumRegister(1, "A")
ebit1 = QuantumRegister(1, "B")
a = ClassicalRegister(1, "a")
b = ClassicalRegister(1, "b")
 
protocol = QuantumCircuit(qubit, ebit0, ebit1, a, b)
 
# Prepare ebit used for teleportation
protocol.h(ebit0)
protocol.cx(ebit0, ebit1)
protocol.barrier()
 
# Alice's operations
protocol.cx(qubit, ebit0)
protocol.h(qubit)
protocol.barrier()
 
# Alice measures and sends classical bits to Bob
protocol.measure(ebit0, a)
protocol.measure(qubit, b)
protocol.barrier()
 
# Bob uses the classical bits to conditionally apply gates
with protocol.if_test((a, 1)):
    protocol.x(ebit1)
with protocol.if_test((b, 1)):
    protocol.z(ebit1)
 
display(protocol.draw(output="mpl"))
```

```python
random_gate = UGate(
    theta=random.random() * 2 * pi,
    phi=random.random() * 2 * pi,
    lam=random.random() * 2 * pi,
)
 
display(array_to_latex(random_gate.to_matrix()))
```

```python
# Create a new circuit including the same bits and qubits used in the
# teleportation protocol.
 
test = QuantumCircuit(qubit, ebit0, ebit1, a, b)
 
# Start with the randomly selected gate on Q
 
test.append(random_gate, qubit)
test.barrier()
 
# Append the entire teleportation protocol from above.
 
test = test.compose(protocol)
test.barrier()
 
# Finally, apply the inverse of the random unitary to B and measure.
 
test.append(random_gate.inverse(), ebit1)
result = ClassicalRegister(1, "Result")
test.add_register(result)
test.measure(ebit1, result)
 
display(test.draw(output="mpl"))
```

```python
result = AerSimulator().run(test).result()
statistics = result.get_counts()
display(plot_histogram(statistics))
```

### Jeu CHSH

```python
def chsh_game(strategy):
    # This function runs the CHSH game, using the strategy (a function
    # from two bits to two bits), returning 1 for a win and 0 for a loss.
 
    # Choose x and y randomly
    x, y = random.randint(0, 2), random.randint(0, 2)
 
    # Use the strategy to determine a and b
    a, b = strategy(x, y)
 
    # Decide if the strategy wins or loses
    if (a != b) == (x & y):
        return 1  # Win
    return 0  # Lose
```

```python
def chsh_circuit(x, y):
    # This function creates a `QuantumCircuit` implementing the quantum
    # strategy described above (including the e-bit preparation).
 
    qc = QuantumCircuit(2, 2)
 
    # Prepare an e-bit
    qc.h(0)
    qc.cx(0, 1)
    qc.barrier()
 
    # Alice's actions
    if x == 0:
        qc.ry(0, 0)
    else:
        qc.ry(-pi / 2, 0)
    qc.measure(0, 0)
 
    # Bob's actions
    if y == 0:
        qc.ry(-pi / 4, 1)
    else:
        qc.ry(pi / 4, 1)
    qc.measure(1, 1)
 
    return qc
```

```python
# Draw the four possible circuits
 
print("(x,y) = (0,0)")
display(chsh_circuit(0, 0).draw(output="mpl"))
 
print("(x,y) = (0,1)")
display(chsh_circuit(0, 1).draw(output="mpl"))
 
print("(x,y) = (1,0)")
display(chsh_circuit(1, 0).draw(output="mpl"))
 
print("(x,y) = (1,1)")
display(chsh_circuit(1, 1).draw(output="mpl"))
```

```python
def quantum_strategy(x, y):
    # This function runs the appropriate quantum circuit defined above
    # one time and returns the measurement results
 
    # Setting `shots=1` to run the circuit once
    result = AerSimulator().run(chsh_circuit(x, y), shots=1).result()
    statistics = result.get_counts()
 
    # Determine the output bits and return them
    bits = list(statistics.keys())[0]
    a, b = bits[0], bits[1]
    return a, b
```

```python
NUM_GAMES = 1000
TOTAL_SCORE = 0
 
for _ in range(NUM_GAMES):
    TOTAL_SCORE += chsh_game(quantum_strategy)
 
print("Fraction of games won:", TOTAL_SCORE / NUM_GAMES)
```

```python
def classical_strategy(x, y):
    # This function implements just one example of an optimal classical
    # strategy for the CHSH game. Other classical strategies can be
    # implemented by changing the bit values assigned to a and b.
 
    # Alice's answer
    if x == 0:
        a = 0
    elif x == 1:
        a = 1
 
    # Bob's answer
    if y == 0:
        b = 1
    elif y == 1:
        b = 0
 
    return a, b
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator
import numpy as np

# Créer un circuit quantique avec 2 qubits
qc = QuantumCircuit(2)

# Appliquer les portes selon le schéma
qc.h(0)           # Porte de Hadamard sur le qubit 0
qc.z(1)           # Porte Z sur le qubit 1
qc.cx(0, 1)       # Porte CNOT (contrôle: qubit 0, cible: qubit 1)

# Afficher le circuit
print("Circuit quantique:")
print(qc.draw())

# Afficher le vecteur d'état
display(qc.draw(output="mpl"))

# Calculer le vecteur d'état final
statevector = Statevector.from_instruction(qc)

# Afficher le vecteur d'état
print("Vecteur d'état final:")
print(statevector)
print("\n")


# Afficher les amplitudes pour chaque état de base
print("Amplitudes des états de base:")
for i, amplitude in enumerate(statevector.data):
    if np.abs(amplitude) > 1e-10:  # Afficher seulement les amplitudes non-nulles
        basis_state = format(i, '02b')
        print(f"|{basis_state}⟩: {amplitude:.4f}")

# Simulation avec mesures
qc_measure = qc.copy()
qc_measure.measure_all()

# Simuler le circuit
simulator = AerSimulator()
job = simulator.run(qc_measure, shots=1000)
result = job.result()
counts = result.get_counts()

print("Résultats des mesures (1000 shots):")
for state, count in sorted(counts.items()):
    print(f"|{state}⟩: {count} fois ({count/10:.1f}%)")

print("\n" + "="*50 + "\n")

# Vérifier quel état de Bell c'est
print("Identification de l'état de Bell:")
print("État obtenu: (1/√2)|00⟩ + (1/√2)|11⟩")
print("Ceci correspond à l'état de Bell |Φ⁺⟩")
```
